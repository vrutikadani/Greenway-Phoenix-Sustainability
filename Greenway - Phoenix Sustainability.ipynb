PART II: DATA ANALYSIS 

```python
# Outlier Detection
sns.boxplot(x='REFUSE', y='NUMBER_OF_CONTAINERS', data=phoenix_data)
plt.title('Box plot: REFUSE vs. NUMBER_OF_CONTAINERS')
plt.xlabel('Refuse')
plt.ylabel('Number of Containers')
plt.show()
```

```python
# Filter out data points with number of containers less than or equal to 20000
filtered_data = phoenix_data[phoenix_data['NUMBER_OF_CONTAINERS'] <= 20000]

# Scatter plot for filtered data
sns.scatterplot(x='NUMBER_OF_CONTAINERS', y='GIS_X_COORDINATE', data=filtered_data)
plt.title('Scatter plot: NUMBER_OF_CONTAINERS vs. GIS_X_COORDINATE (Number of Containers <= 20000)')
plt.xlabel('Number of Containers')
plt.ylabel('GIS X Coordinate')
plt.show()
```

```python
# Drop non-numeric columns
numeric_columns = phoenix_data.select_dtypes(include=['number']).columns
numeric_data = phoenix_data[numeric_columns]

# Calculate correlation matrix
corr_matrix = numeric_data.corr()

# Plot correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Histogram for numerical variables
plt.hist(phoenix_data['NUMBER_OF_CONTAINERS'], bins=20)
plt.title('Histogram of NUMBER_OF_CONTAINERS')
plt.xlabel('Number of Containers')
plt.ylabel('Frequency')
plt.show()
```
